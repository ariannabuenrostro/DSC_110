# Functions

This line is assigning a function to a variable. This can be helpful when needing to use a function multiple times without having to write out the whole function.

```python
temp=(4*9/5)+32
print(temp)
```

```python
temp=(-6*5/9)+32
print(temp)
```

These lines of code are defining a function that converts the temperature in celsius to farenheit. The name of this function will be "c_temp". In the end of the function the user will receive the conversion to the temp in farenheit with the whole function (c_temp*(9/5))+32. Then the answer will be returned to the user with the "return" command. By saying that "f_temp = c_to_f(4)", it means the degrees in farenheit the user will receive is equal to the function that converts the temp in celsius to the temp in farenheit at 4 degrees celsius. By making functions like this so compact and easy to read it can make an otherwise lengthy code consice and efficient.

```python
def c_to_f(c_temp):
  f_temp=(c_temp*(9/5))+32
  return f_temp
```

```python
f_temp = c_to_f(4)
print (f_temp)
```

Loops

These lines of code are assigning values to a list named "my_temp_list_c". Then to make sure the function created earlier that converts the temperature in celsius to farenheit works some values were inputted there as well. The third query is pulling the data from the list named "my_temp_list_c" and inputting it into the function "c_to_f". This allows for more function and less lines written.

```python
my_temp_list_c=[0,10,20,30,40]
```

```python
c_to_f(0)
c_to_f(10)
```

These lines are assigning placeholders for two values that the user inputs into a function named "calculate_bmi" that calculates an individuals body mass index. Then it assigns "bmi" as a variable that will equal the function that calculates the bmi and then return the result to the user. The second line is inserting the values into the bmi function and displaying the answer back to the function

```python
def calculate_bmi(weight_kg, height_m):
  bmi= weight_kg / (height_m**2)
  return bmi
```
