```python
temp=(4*9/5)+32
print(temp)
```

```python
temp=(-6*5/9)+32
print(temp)
```

```python
def c_to_f(c_temp):
  f_temp=(c_temp*(9/5))+32
  return f_temp
```

```python
f_temp = c_to_f(4)
print (f_temp)
```

Formatted string literals

```python
print(f'the temperature now is {c_to_f(0)}F')
```

```python
print(f'the temperature now is {c_to_f(35)}F')
```

Loops

```python
my_temp_list_c=[0,10,20,30,40]
```

```python
c_to_f(0)
c_to_f(10)
```

```python
for c_temp in my_temp_list_c:
  print(c_to_f(c_temp))
```

```python
def calculate_bmi(weight_kg, height_m):
  bmi= weight_kg / (height_m**2)
  return bmi
```

```python
bmi= calculate_bmi(70,1.75)
print(bmi)
```
